```python
#Displaying basic data about our league
teams_info = r.json()[0]
teams_info
```

```python
#Putting teams data into a dataframe
# for team in teams_info["teams"]:
#     print(team)
    
teams_lst = [team for team in teams_info["teams"]]
teams_df = pd.DataFrame(teams_lst)
teams_df["owners"] = teams_df["owners"].str[0]
teams_df
```

```python
members_lst = [member for member in teams_info["members"]]
members_df = pd.DataFrame(members_lst)
members_df.columns = ["displayName", "owners", "isLeagueManager"]
members_df
```

```python
teams_df = teams_df.merge(members_df, on="owners")
teams_df
```

```python
teams_df = teams_df.drop("isLeagueManager", axis=1)
teams_df
```

```python
#Pulling mMatchup data
r = requests.get(url, cookies={"swid": swid, "espn_s2": espn_s2}, params={"view": "mMatchup"})

matchup_json = r.json()[0]
```

```python
game_lst = [[game['matchupPeriodId'],
            game['home']['teamId'], game['home']['totalPoints'],
            game['away']['teamId'], game['away']['totalPoints']
           ] for game in matchup_json['schedule']]

games_df = pd.DataFrame(game_lst, columns=['Week', 'Team1', 'Score1', 'Team2', 'Score2'])
games_df["Type"] = ["Regular" if w <=14 else "Playoff" for w in games_df["Week"]]
games_df.head()
```

```python
games_df["Winner"] = np.where(games_df["Score1"] > games_df["Score2"], games_df["Team1"], games_df["Team2"])
games_df.head()
```
